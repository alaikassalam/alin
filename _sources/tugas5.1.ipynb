# Determinan Matriks 4x4 dengan Ekpansi Kofaktor 

**1. langkah pertama buat matriksnya terlebih dahulu dengan ordo 4x4**

   
$\begin{matrix}
A = \begin{bmatrix}
    2 & 3 & 4 & 5\\
    1 & 1 & 1 & 1\\
    3 & 2 & 2 & 2\\
    3 & 2 & 5 & 4
  \end{bmatrix}
\end{matrix}
$

**2. selanjutnya tulis rumus determinan disini, mengambil baris ke-2**


$ det(A) = a_{21} c_{21} + a_{22} c_{22}  + a_{23} c_{23} + a_{24} c_{24} $  

**3. selanjutnya tentukan kofaktor nya disini dari matriks A**


$c_{21} = (-1)^{2+1} det(M_{21})$  
$c_{22} = (-1)^{2+2} det(M_{22})$  
$c_{23} = (-1)^{2+3} det(M_{23})$  
$c_{24} = (-1)^{2+4} det(M_{24})$

**4. setelah itu kita akan cari satu-persatu determinan minor tersebut dari matriks A**

**5. **pertama : $M_{21}$****


$\begin{matrix}
  det(M21) = \begin{bmatrix}
    3 & 4 & 5 \\
    2 & 2 & 2 \\
    2 & 5 & 4
  \end{bmatrix}
\end{matrix}
$  


selanjutnya tulis rumus determinan, dan mengambil baris ke-2  
$det(M21) = a_{21}c_{21} + a_{22}c_{22} + a_{23}c_{23}$   


selanjutnya tentukan kofaktor dari matriks yang $M_{21}, M_{22}, M_{23}$  
* $c_{21} = (-1)^{2 + 1} det(KM21)$   

$
\begin{matrix}
  det(KM21) = \begin{bmatrix}
    4 & 5 \\
    5 & 4
  \end{bmatrix}
\end{matrix}
$  
$det(KM21) = 4.4 - 5.5 = 16 - 25 = -9$  


hasil 
$c_{21} = -1*-9$  


* $c_{22} = (-1)^{2+2}det(KM22)$   

$
\begin{matrix}
  det(KM22) = \begin{bmatrix}
    3 & 5 \\
    2 & 4
  \end{bmatrix}
\end{matrix}
$  
$det(KM22) = 3.4 - 2.5 = 12 - 10 = 2$  


hasil
$c_{22} = -1*2$  


* $c_{23} = (-1)^{2+3}det(KM23)$  
 
$
\begin{matrix}
  det(KM23) = \begin{bmatrix}
    3 & 4 \\
    2 & 5
  \end{bmatrix}
\end{matrix}
$  
$det(KM23) = 3.5 - 2.4 = 15 - 8 = 7$  


hasil
$c_{23} = -1*7$  


**hasil akhir dari $M_{21}$**  
$det(M21) = a_{21}c_{21} + a_{22}c_{22} + a_{23}c_{23}$  
$det(M21) = 2.(-1).(-9) + 2.1.2 + 2.(-1).7$  
$det(M21) = 18 + 4 - 14 = 8$  


dibuktikan dengan menggunakan numpy python 

```python
import numpy as np
M21 = np.array([[3,4,5], [2,2,2], [2,5,4]])
detM21 = np.linalg.det(M21)
print(round(detM21))
```

**9. **langkah terakhir mencari determinan A****  


$det(M_{21}) = 8$  
$det(M_{22}) = 17$  
$det(M_{23}) = 0$  
$det(M_{24}) = -15$  

selanjutnya subtitusi nilai determinan dari matriks minornya  

$c_{21} = (-1)^{2+1}det(M_{21}) = -1.8$  
$c_{22} = (-1)^{2+1}det(M_{22}) = 1.17$  
$c_{23} = (-1)^{2+1}det(M_{23}) = -1.0$  
$c_{24} = (-1)^{2+1}det(M_{24}) = -1.-15$  

$a_{21}c_{21} + a_{22}c_{22} + a_{23}c_{23} + a_{24}c_{24} =$
$1.(-1).8 + 1.1.17 + 1.(-1).0 + 1.(-1).(-15) =$
$-8 + 17 + 0 + 15 = 4$  


```python
import numpy as np
A = np.array([[2,3,4,5], [1,1,1,1], [3,2,2,2], [3,2,5,4]])
detA = np.linalg.det(A)
print("A: \n", A)
print()
print("Determinan A = ", round(detA))
```
