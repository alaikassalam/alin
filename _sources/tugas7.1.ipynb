# Implementasi Transformasi Linier


Pertama, kita menuliskan bentuk umum matriks $ A $ yang memetakan vektor standar:

$$
A = \begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
$$

Tugas kita adalah menemukan $ a, b, c, d $ sehingga syarat-syarat berikut terpenuhi.

1. Berdasarkan informasi pertama:

$$
A \begin{pmatrix}
2 \\
0
\end{pmatrix} = \begin{pmatrix}
-2 \\
0
\end{pmatrix}
$$

Hal ini memberikan dua persamaan:

$$
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\begin{pmatrix}
2 \\
0
\end{pmatrix} = \begin{pmatrix}
2a \\
2c
\end{pmatrix} = \begin{pmatrix}
-2 \\
0
\end{pmatrix}
$$

Dari sini, kita dapatkan:

$$
\begin{cases}
2a = -2 \implies a = -1 \\
2c = 0 \implies c = 0
\end{cases}
$$

2. Berdasarkan informasi kedua:

$$
A \begin{pmatrix}
1 \\
2
\end{pmatrix} = \begin{pmatrix}
-2 \\
-2
\end{pmatrix}
$$

Hal ini memberikan dua persamaan lagi:

$$
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\begin{pmatrix}
1 \\
2
\end{pmatrix} = \begin{pmatrix}
a + 2b \\
c + 2d
\end{pmatrix} = \begin{pmatrix}
-2 \\
-2
\end{pmatrix}
$$

Dari sini, kita dapatkan:

$$
\begin{cases}
a + 2b = -2 \implies -1 + 2b = -2 \implies 2b = -1 \implies b = -\frac{1}{2} \\
c + 2d = -2 \implies 0 + 2d = -2 \implies d = -1
\end{cases}
$$

Sehingga nilai-nilai dari $ a, b, c, d $ adalah:

$$
a = -1, \; b = -\frac{1}{2}, \; c = 0, \; d = -1
$$

Dengan demikian, matriks transformasi yang kita peroleh adalah:

$$
A = \begin{pmatrix}
-1 & -\frac{1}{2} \\
0 & -1
\end{pmatrix}
$$

Setelah menentukan nilai $ a, b, c, d $, kita mendapatkan:

$$
A = \begin{pmatrix}
-1 & -\frac{1}{2} \\
0 & -1
\end{pmatrix}
$$
